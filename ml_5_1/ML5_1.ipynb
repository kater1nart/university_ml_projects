```python
data_left = {'key': ['A', 'B', 'C', 'D'], 'value_left': [10, 20, 15, 25]}
df_left = pd.DataFrame(data_left)
df_left.head()
```

```python
data_right = {'key': ['B', 'D', 'E', 'F'], 'value_right': [30, 40, 35, 45]}
df_right = pd.DataFrame(data_right)
df_right.head()
```

```python
import matplotlib.pyplot as plt
```

```python
for column in df_left.columns:
    if df_left[column].dtype in ['int64', 'float64']:
        plt.figure()
        df_left[column].hist()
        plt.title(f'Распределение столбца "{column}" в df_left')
        plt.xlabel(column)
        plt.ylabel('Частота')
        plt.show()
    else:
        plt.figure()
        df_left[column].value_counts().plot(kind='bar')
        plt.title(f'Распределение столбца "{column}" в df_left')
        plt.xlabel(column)
        plt.ylabel('Количество')
        plt.show()
```

```python
for column in df_right.columns:
    if df_right[column].dtype in ['int64', 'float64']:
        plt.figure()
        df_right[column].hist()
        plt.title(f'Распределение столбца "{column}" в df_right')
        plt.xlabel(column)
        plt.ylabel('Частота')
        plt.show()
    else:
        plt.figure()
        df_right[column].value_counts().plot(kind='bar')
        plt.title(f'Распределение столбца "{column}" в df_right')
        plt.xlabel(column)
        plt.ylabel('Количество')
        plt.show()
```

```python
df_merged_horizontal = pd.merge(df_left, df_right, on='key', how='inner')
df_merged_horizontal.head()
```
