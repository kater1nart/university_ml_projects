## Заполнение пропусков

```python
sns.heatmap(prsa_data.isnull(), yticklabels=False, cbar=False)
```

```python
undef = prsa_data.isnull().sum(axis=1)
undef[undef >= 2]
```

```python
prsa_data = prsa_data.drop(undef[undef >= 2].index, axis=0)
```

```python
prsa_data.isnull().sum()
```

```python
prsa_data.SO2 = prsa_data.SO2.fillna(prsa_data.PRES.mean())
```

```python
filler = prsa_data.O3[prsa_data.O3.isna()]
```

```python
filler = prsa_data.O3[~prsa_data.O3.isna()].sample(n=len(filler)).set_axis(filler.index)
```

```python
prsa_data.O3 = prsa_data.O3.fillna(filler)
```

## Округление атрибутов

```python
prsa_data['O3'] = np.array(np.round((prsa_data['O3'])), dtype='int')
prsa_data.head()
```

## Логарифмирование атрибутов

```python
prsa_data.SO2 = np.log(prsa_data.SO2)
```

```python
def trim_outliers_by_percentiles(df, columns, lower_perc=1, upper_perc=99):
    df_trimmed = df.copy()
    for col in columns:
        # Вычисляем границы
        lower_bound = df[col].quantile(lower_perc/100)
        upper_bound = df[col].quantile(upper_perc/100)

        # Обрезаем значения
        df_trimmed[col] = df[col].clip(lower_bound, upper_bound)

        # Выводим статистику
        n_outliers = ((df[col] < lower_bound) | (df[col] > upper_bound)).sum()
        print(f"{col}: обрезано {n_outliers} значений ({lower_perc}%-{upper_perc}%: {lower_bound:.2f}-{upper_bound:.2f})")

    return df_trimmed

# Применяем к числовым столбцам
numeric_cols = ['SO2', 'NO2', 'CO', 'O3', 'PRES', 'WSPM']
prsa_trimmed = trim_outliers_by_percentiles(prsa_data, numeric_cols)
```

Так как у NO2 всего 2% проусков, заполни их медианной с группировкой по AQI Label

```python
prsa_data['NO2'] = prsa_data.groupby('AQI Label')['NO2'].transform(
    lambda x: x.fillna(x.median())
)
```

У CO также мало пропусков всего 3% и можно было бы заполнить так же медианным значением, но для другого способа создам вспомогательную модель для предсказания пропусков

```python
from sklearn.ensemble import RandomForestRegressor

train_data = prsa_data.dropna(subset=['CO'])
test_data = prsa_data[prsa_data['CO'].isnull()]

model = RandomForestRegressor(n_estimators=100, random_state=42)
features = ['SO2', 'NO2', 'O3', 'PRES', 'WSPM', 'IS_RAIN']
model.fit(train_data[features], train_data['CO'])

prsa_data.loc[prsa_data['CO'].isnull(), 'CO'] = model.predict(test_data[features])
```

```python
def analyze_feature(df, col):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
    sns.histplot(df[col], kde=True, ax=ax1)
    sns.boxplot(x=df[col], ax=ax2)
    plt.suptitle(f'Анализ распределения: {col}')
    plt.show()
analyze_feature(prsa_data, 'SO2')
analyze_feature(prsa_data, 'NO2')
analyze_feature(prsa_data, 'CO')
analyze_feature(prsa_data, 'O3')
analyze_feature(prsa_data, 'PRES')
analyze_feature(prsa_data, 'WSPM')
```

2. NO2

RobustScaler (лучший выбор для данных с тяжелыми хвостами)

MinMaxScaler (диапазон четко ограничен)

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler

standard_scaler = StandardScaler()
minmax_scaler = MinMaxScaler()
robust_scaler = RobustScaler()
```

```python
prsa_data['SO2_normalized'] = robust_scaler.fit_transform(prsa_data[['SO2']])
prsa_data['NO2_normalized'] = standard_scaler.fit_transform(prsa_data[['NO2']])
prsa_data['CO_normalized'] = robust_scaler.fit_transform(prsa_data[['CO']])
prsa_data['O3_normalized'] = minmax_scaler.fit_transform(prsa_data[['O3']])
prsa_data['PRES_normalized'] = standard_scaler.fit_transform(prsa_data[['PRES']])
prsa_data['WSPM_normalized'] = minmax_scaler.fit_transform(prsa_data[['WSPM']])
```

```python
aqi_label_mapping = {
    'Severely Polluted': 0,
    'Heavily Polluted': 1,
    'Moderately Polluted': 2,
    'Lightly Polluted': 3,
    'Slightly Polluted': 4,
    'Good': 5,
    'Excellent': 6
}
prsa_data['AQI_Score'] = prsa_data['AQI Label'].map(aqi_label_mapping)
```

```python
numeric_features = ['SO2', 'NO2', 'CO', 'O3', 'PRES', 'WSPM', 'IS_RAIN', 'AQI_Score']

plt.figure(figsize=(12, 8))
sns.heatmap(prsa_data[numeric_features].corr(),
            annot=True,
            cmap='coolwarm',
            vmin=-1, vmax=1,
            linewidths=0.5)
plt.title('Корреляционная матрица числовых признаков')
plt.show()
```

Из матрицы видно, что AQI_Score имеет следующие значимые корреляции:

Наибольшая отрицательная корреляция:

NO2 (-0.28): Чем выше диоксид азота, тем хуже качество воздуха (ожидаемо).

CO (-0.21): Высокий уровень CO также снижает AQI.

Незначимые корреляции:

O3 (0.039): Озон почти не влияет на AQI в этих данных.

WSPM (0.26): Слабая положительная связь — возможно, ветер рассеивает загрязнители.

Вывод:
Основные "враги" качества воздуха — NO2 и CO (промышленные/транспортные выбросы).
