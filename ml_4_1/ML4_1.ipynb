```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('heart.csv')
data.head()
```

```python
y = data["output"]
x = data.drop("output", axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, random_state=42)
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
```

```python
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42),
    "SVM": SVC(random_state=42),
    "K-Nearest Neighbors": KNeighborsClassifier(),
    "Naive Bayes": GaussianNB(),
    "MLP Classifier": MLPClassifier(random_state=42, max_iter=1000),
}
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
results = {}
for name, model in models.items():
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    results[name] = {
        "Accuracy": accuracy_score(y_test, y_pred),
        "Precision": precision_score(y_test, y_pred),
        "Recall": recall_score(y_test, y_pred),
        "F1": f1_score(y_test, y_pred),
    }
    print(f"\n{name} classification report:\n{classification_report(y_test, y_pred)}")
```

```python
results_df = pd.DataFrame(results).T
print(results_df)
```

**Лучшие модели:** Logistic Regression, Naive Bayes и MLP Classifier показывают наилучшие результаты. Они обеспечивают высокую точность, полноту и F1-меру.
**Худшие модели:** Decision Tree, SVM и K-Nearest Neighbors показывают более низкие результаты. Возможно, для этих моделей требуется дополнительная настройка параметров или использование других методов обработки данных.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
```

```python
wine = load_wine()
wine
```

```python
wine_df = pd.DataFrame(data=wine['data'], columns=wine['feature_names'])
wine_df['target'] = wine['target']
wine_df.head()
```

```python
y = wine_df['target']
x = wine_df.drop('target', axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, random_state=42)
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
```

```python
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42),
    "SVM": SVC(random_state=42),
    "K-Nearest Neighbors": KNeighborsClassifier(),
    "Naive Bayes": GaussianNB(),
    "MLP Classifier": MLPClassifier(random_state=42, max_iter=1000),
}
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
results = {}
for name, model in models.items():
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    results[name] = {
        "Accuracy": accuracy_score(y_test, y_pred),
        "Precision": precision_score(y_test, y_pred, average='weighted'),
        "Recall": recall_score(y_test, y_pred, average='weighted'),
        "F1": f1_score(y_test, y_pred, average='weighted'),
    }
    print(f"\n{name} classification report:\n{classification_report(y_test, y_pred)}")
```

```python
results_df = pd.DataFrame(results).T
print(results_df)
```

SVM и Naive Bayes — лучшие модели

K-Nearest Neighbors показал слабые результаты

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing
```

```python
california = fetch_california_housing()
california
```

```python
california_df = pd.DataFrame(data=california['data'], columns=california['feature_names'])
california_df['target'] = california['target']
california_df.head()
```

```python
y = california_df['target']
x = california_df.drop('target', axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, random_state=42)
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, mean_absolute_percentage_error, median_absolute_error, max_error
```

```python
models = {
    "Linear Regression": LinearRegression(),
    "Decision Tree Regressor": DecisionTreeRegressor(random_state=42),
    "Random Forest Regressor": RandomForestRegressor(random_state=42),
    "Support Vector Regressor": SVR(),
    "K-Nearest Neighbors Regressor": KNeighborsRegressor(),
    "MLP Regressor": MLPRegressor(random_state=42, max_iter=1000),
}
```

```python
results = {}
for name, model in models.items():
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)

    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    rmse = mean_squared_error(y_test, y_pred)
    mape = mean_absolute_percentage_error(y_test, y_pred)
    medae = median_absolute_error(y_test, y_pred)
    max_er = max_error(y_test, y_pred)

    results[name] = {
        "MSE": mse,
        "MAE": mae,
        "R2": r2,
        "RMSE": rmse,
        "MAPE": mape,
        "MedAE": medae,
        "Max Error": max_er,
    }
```

```python
results_df = pd.DataFrame(results).T
print(results_df)
```

Random Forest Regressor — лучшая модель

Support Vector Regressor и K-Nearest Neighbors Regressor показали плохие результаты
